```python
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
import requests
```

```python
csv = pd.read_csv("delhi.csv")
data = csv.copy()
data
```

```python
reqData = data.drop(["date_time","moonrise","moonset","sunrise","sunset","cloudcover","precipMM","moon_illumination",
          "totalSnow_cm","sunHour","uvIndex.1","uvIndex"],axis=1)


reqData.to_csv("Delhi_1.csv")
```

```python
li = list()
for i in range(0,len(reqData),24):
    li.append(list(reqData[i:i+24].mean()))
```

```python
new = pd.DataFrame(li,columns=list(reqData.columns))
```

```python
new.to_csv("delhi_2_day.csv")
```
